Analisis Exploratorio de Datos del CIC-IDS2017


```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

Dataset original sin procesar de uno de los archivos .csv

```python
df = pd.read_csv("../DataSetCICIDS2017/MachineLearningCVE/Friday-WorkingHours-Afternoon-DDos.pcap_ISCX.csv")
df.head()
df.info()
print(df[' Label'].value_counts())
```

Carga, limpieza y visualizaciones iniciales del dataset combinado y procesado

```python

df = pd.read_csv("../Datos_procesados_combinados/dataset_combinado.csv")
df.head()
df.info()
df['Label'].value_counts()
```

Flow Duration

```python
plt.figure(figsize=(8, 4)) #Ancho=8 y Alto=4
df['Flow Duration'].hist(bins=50)
plt.title("Distribucion de la duracion de flujos")
plt.xlabel("Duracion de Flujos")
plt.ylabel("Frecuencia")
plt.tight_layout()
plt.savefig("../Notebooks/histograma_duracion_flujos.png")
```

Diagrama de Bloques

```python
plt.figure(figsize=(6, 4))
df.boxplot(column='Flow Bytes/s')
plt.title("Diagrama de bloques de Bytes/s")
plt.tight_layout()
plt.savefig("../Notebooks/diagramabloques_bytes.png")
```

Matriz de Correlacion

```python
import seaborn as sns
plt.figure(figsize=(11, 11))
sns.heatmap(df.corr(), center=0)
plt.title("Matriz de correlacion")
plt.tight_layout()
plt.savefig("../Notebooks/matrizcorrelacion.png")
plt.title("Matriz de correlacion")
plt.tight_layout()
plt.savefig("../Notebooks/matrizcorrelacion.png")
```
